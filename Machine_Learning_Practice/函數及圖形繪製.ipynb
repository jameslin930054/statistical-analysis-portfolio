$f(x) = \sin(x) + \cos(x) $

```python
import numpy as np
import matplotlib.pyplot as plt
 
# The codes below demonstrate the ticks and labels in x axis
x = np.linspace(-3*np.pi, 3*np.pi,200) #x的範圍
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2
fig, ax = plt.subplots(1)
ax.plot(x, y1, linestyle='--',label = 'sin(x)',color='b')
ax.plot(x, y2, linestyle='--',label = 'cos(x)',color='g')
ax.plot(x, y3,label = 'sin(x)+cos(x)',color='r')
ax.set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi) #x座標軸寬度
ax.set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color = 'b') #標示坐標軸
ax.set_ylim([-3, 3]) #y座標軸寬度
ax.grid(True) #顯示格線 
ax.annotate('$f(x)=\sqrt{2} $', \
    xy=(-1, 1.5), xytext=(0.1, 1.7), \
        arrowprops=dict(arrowstyle='->', \
            connectionstyle='arc3, rad=0.3'), fontsize =14) 
ax.annotate('$f(x)=-\sqrt{2} $', \
    xy=(-1,-1.4), xytext=(-0.1, -2), \
        arrowprops=dict(arrowstyle='->', \
            connectionstyle='arc3, rad=-0.5'), fontsize =14) 
plt.axhline(y=-2**(0.5),linestyle='--', xmin=-3*np.pi, xmax=3*np.pi)
plt.axhline(y=2**(0.5),linestyle='--', xmin=-3*np.pi, xmax=3*np.pi)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend() 
plt.show()
```

$f(x) = \frac{1-e^{-2x}}{1+e^{-2x}}$

```python
import matplotlib.pyplot as plt
import numpy as np
fx_name = r'$f(x) = \frac{1-e^{-2x}}{1+e^{-2x}}$'
n = 100
x = np.linspace(-3, 3, n) 
y = (1-np.exp(-2*x))/(1+np.exp(-2*x))
plt.plot(x, y,label = fx_name,color='r',linewidth =3)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='lower right')
plt.show()  
```

$f(x) = \sqrt[3]{\frac{4-x^3}{1+x^2}}$

```python
import matplotlib.pyplot as plt
import numpy as np
fx_name = r'$f(x) = \sqrt[3]{\frac{4-x^3}{1+x^2}}$'
n = 100
x = np.linspace(-3, 3, n)
y1 = 4-(x**3)
y2 = 1+(x**2)
y3 = np.cbrt(y1/y2)
y4 = (y1/y2)**(1/3)
plt.plot(x, y3, label=fx_name, linewidth=3)
plt.plot(x, y4, label=fx_name, linestyle='--', alpha=0.7, linewidth=3)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper right')
plt.show()
```

$f(x) = \frac{1}{2\sqrt{2\pi}} e^{-\frac{(x-1)^2}{8}}$
